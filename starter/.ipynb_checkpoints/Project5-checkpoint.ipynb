Explain three principles of effective data communication from your lecture material: 
1) Simplicity - visualizations should be simple and easy to understand, not cluttered. Labels should be clear and concise. Don't overwhelm the audience with too much information. 
2) The context needs to be clear. For example, if the audience is Samsung and you are presenting sales data from t-mobile stores, the data presented should be relevant to the audience. 
3) Find ways to connect to the audience, to make the data relatable to them. If it is too abstract they might not follow or lose interest. 

2) 

Explain the following concepts as they relate to visual perception: order, hierarchy, relationships, and convention. How does an understanding of these concepts help you create better visualizations?

Order: Order refers to the arrangement of the elements in the visualization. This could be the order of categories on a bar graph, or what is chosen as the x and y axis on a graph intended to show correlations. The clarity and impact of the visualization can be affected by the order of the elements. 

Hierarchy: Hierarchy refers to the arrangement of elements in a visualization to indicate which one is emphasized more. Size, position, and color are some ways to emphasize certain elements over others. 

Relationships: Usually we study data to find relationships between the variables. The relationships between data points or categories could be important here. 

Convention: The standard practices used in visualization, such as standard chart types, scales (percent, logarithmic). 

Understanding these principles can make your visualizations more clear, impactful, and memorable. 

3) 

What type of graph would be best for 
- comparison between values: a bar graph or line graph. 
- comparison to the whole : a pie chart or stacked bar/column chart. 
- change over time: a line graph. 
- ranking data: a bar graph 
- for correlation: a scatter plot 
- geographical charts: a map
- measuring a target: line graph, bar graph, pie chart, scatter plot, or area graph. 
- showing outliers: box plot, scatter plot, dot plot, or line plot. 

Using the appropriate type of graph will be useful to make the communication more effective. 

4) 

Consider the following quote:

“It is easy to lie with statistics. It is hard to tell the truth without it.” - Andrejs Dunkels

What do you think is meant by this statement?

Provide three examples of misleading graphs. Explain what is misleading about the graph in your example and what should be changed to make the graph objective and accurate.

This statement means that statistics can be used to manipulate data. A biased or misleading picture can be presented in some cases. It can also be used to reveal the truth. 

Example 1: If the y-axis starts at a very large value rather than 0, it can make the differences in bars on a bar graph appear larger than they really are. 

Example 2: Pie charts: If the small slices are very small, the larger portion can appear larger than it really is. 
Example 3: Line Graphs: If a line graph uses a curve that smoothes out the fluctuations in the data, it can give the impression that the data is more consistent than it actually is.

5) What is “visualization clutter?”
What are the main components of a graph?
What are three techniques you learned to make data visualizations more clear?
How can the use of color affect the way your visualizations are understood?


Visualization clutter is the overuse or misuse of visual elements. The main components of a graph are: 

The title, x-axis, y-axis, data, labels, legend, title, labels. 

You can make data visualization more clear by simplifying the graph (removing unnecessary elements and information from the graph), highlighting important information (using colors, patterns, or symbols to emphasize import data points or trends, and interactivity (allowing the audience to interact with the graph by zooming in, panning, or highlighting specific data points). 

Color can distinguish data points or elements on the graph, highlight important information, and convey a message. 

Sometimes color can be confusing. For instance if you used red to indicate positive values and blue to indicate negative values it might be confusing since these colors are usually associated with the opposite meaning.

```python
df_supermarket = pd.read_csv('supermarket.csv')
```

```python
df_supermarket.head()
```

```python
'''Top 10 performing stores, 
   Top 5 performing areas, 
    Average daily customers overall'''


# Find the top 10 performing stores 
 

df_supermarket.sort_values(['store_sales'], ascending=False).head(10)


```

```python
sales = df_supermarket.groupby("store_area")["store_sales"].sum()
```

```python
top_areas = sales.sort_values(ascending=False).head(5)
```

```python
print(top_areas)
```

```python
avg_customers = df_supermarket['daily_customer_count'].mean()
```
